# Исследование данных о российском кинопрокате

**Описание данных** 

Таблица `mkrf_movies`:
- `title` — название фильма;
- `puNumber` — номер прокатного удостоверения;

- `puNumber` — номер прокатного удостоверения;

```python
def describe_column(df, column):
    print(color.BOLD + 'Характерные значения:' + color.END, df[column].describe(percentiles=[.01, .05, .20, .80, .95, .99]), sep='\n', end='\n\n')
    print(color.BOLD + 'Колл-во пропусков:' + color.END, df[column].isna().sum(), end='\n\n')
    print(color.BOLD + 'Колл-во уникальных значений:' + color.END, df[column].nunique(), end='\n\n')
    
    if df[column].dtype == 'int' or df[column].dtype == 'float' and df[column].dtype != 'O':  
        q1 = df[column].quantile(0.25)
        q3 = df[column].quantile(0.75)
        lqr =  q3 - q1
        high_outlier = q3 + 1.5 * lqr
        low_outlier = q1 - 1.5 * lqr

        print(color.BOLD + 'Нижняя граница выбросов (Q1 - 1.5*lQR):' + color.END, round(low_outlier, 2), end='\n\n')
        print(color.BOLD + 'Верхняя граница выбросов (Q3 + 1.5*lQR):' + color.END, round(high_outlier, 2), end='\n\n')
```

### Преобразование типов данных в столбцах

При перовом знакомстве с данными мы заметили, что столбцы `show_start_date`, и `ratings` требуют преобразования типа данных.

#### Столбец 'show_start_date'

Преобразуем тип данных в столбце `show_start_date` в  `datetime`. В значениях оставим только дату, отбросив часы, минуты, и секунды.

```python
data['show_start_date'] = pd.to_datetime(data['show_start_date'], format='%Y-%m-%d').dt.normalize()
data['show_start_date'].head(3)
```

#### Столбец 'ratings'

Проверим уникальные значения в столбце на предмет наличия ошибок.

```python
data['ratings'].unique()
```

В столбце присутсвуют 2 категории данных:
- Оценка по десятибальной шкале (от 0 до 10)
- Оценка в процентой шкале (от 0 до 100%)

Приведем все данные к единому виду. Преобразуем процентные оценки в десятибальные.

```python
# Индексы строк, которые содержат '%' в значениях, запишем в percent_data_indexes
percent_data_indexes = data.loc[data['ratings'].str.contains('%').fillna(False)].index
```

```python
# Убираем символ '%' из значений
data.loc[data['ratings'].index.isin(percent_data_indexes), 'ratings'] = \
data.loc[data['ratings'].index.isin(percent_data_indexes), 'ratings'].str.strip('%')
```

```python
# Преобразуем все значения к типу float
data['ratings'] = data['ratings'].astype(float)
```

```python
# Преобразуем процентные оценки в десятибальные, поделив значения на 10 
data.loc[data['ratings'].index.isin(percent_data_indexes), 'ratings'] = \
data.loc[data['ratings'].index.isin(percent_data_indexes), 'ratings'] / 10
```

Проверим, что преобразование прошло корректно.

```python
data['ratings'].unique()
```

#### Столбец 'age_restriction'

Проверим уникальные значения в столбце на предмет наличия ошибок.

```python
data['age_restriction'].unique()
```

Сократим запись до минимального возраста.

```python
age_shorten = {
    '«18+» - запрещено для детей': '18+',
    '«12+» - для детей старше 12 лет': '12+',
    '«16+» - для детей старше 16 лет': '16+',
    '«6+» - для детей старше 6 лет': '6+',
    '«0+» - для любой зрительской аудитории': '0+'
}

data.replace({'age_restriction': age_shorten}, inplace=True)
data['age_restriction'].unique()
```

**Выводы**

- Значения в столбце `show_start_date` привели к типу `datetime`, отбросив временную часть.
- Значения в столбце `ratings` привели к типу `float`. Скорректировали ошибки, связанные с разным форматом значений оценок (десятибальные и процентные).
- Сократили запись значений в столбце `age_restriction` с длинной: '«18+» - запрещено для детей', до более короткой: '18+'.

Видим, что во многих случаях категориальные столбцы объединяет одна проблема - наличие лишних проблов (и других символов). Избавимся от этох символов. Также приведем данные в разных столбцах к единому стилю написания.

```python
# Уберем лишние пробелы и символы
category_columns = ['title', 'type', 'production_country', 'film_studio', 'director', 'producer']
for column in category_columns:
    data[column] = data[column].str.strip(""" .,_-""")

# Названия стран приведем к стилю 'title', когда каждое новое слово пишеся с большой буквы
# а все промежуточные символы с маленькой. Однако в названиях стран присутствуют аббревиатуры,
# поэтому их нам придется исправить дополнительно
data['production_country'] = data['production_country'].str.title()
country_shorten_names = ['Ссср', 'Сша', 'Оаэ', 'Кнр', 'Юар', 'Внр', 'Гдр']
for country in country_shorten_names:
    data['production_country'] = data['production_country'].str.replace(country, country.upper())

# Преобразуем столбцы в соответствии с предполагаемым написанием их значений
# (доп. пояснения при необходимости нах-ся в выводах)
data['type'] = data['type'].str.capitalize()
data['director'] = data['director'].str.title()
data['producer'] = data['producer'].str.title()
```

Приведем в порядок значения столбца `production_country`, чтобы каждая старна была перечислена через запятую без лишних символов.

```python
# Функция убирает некорректные разделители
def country_names_correction(value):
    value_list = value.strip().split('-')
    value_list = list(filter(None, [i.strip() for i in value_list]))
    value_str = ', '.join(str(i) for i in value_list)
    return value_str

data['production_country'] = data['production_country'].apply(country_names_correction)
```

#### Изучение количества фильмов в прокате по годам

```python
# Посчитаем долю фильмов с указанной информацией о прокате в кинотеатрах
round(1 - data['box_office'].isna().mean(), 2)
```

Теперь **по каждому году** посчитаем количество фильмов с указанной информацией о прокате в кинотеатрах, и без нее.

```python
films_per_year = {}

for year in sorted(data['year'].unique()):
    films_per_year[year] = [data.loc[(data['box_office'].isna() == False) & (data['year'] == year), 'year'].count(), \
                            data.loc[(data['box_office'].isna()) & (data['year'] == year), 'year'].count()]
                        
            
films_per_year
```

```python
# Отобразим полученные данные на гистограмме
years = list(films_per_year.keys())
films_counts = {
    'Данные о прокате известны': np.array([i[0] for i in list(films_per_year.values())]),
    'Данные о прокате неизвестны': np.array([i[1] for i in list(films_per_year.values())]),
}
width = 0.6  # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots(figsize=(12, 5))
bottom = np.zeros(10)

for year, film_count in films_counts.items():
    p = ax.bar(years, film_count, width, label=year, bottom=bottom)
    bottom += film_count

    ax.bar_label(p, label_type='center')

ax.set_title('Сколько фильмов выходило в прокат каждый год')
ax.legend()

plt.xlabel("Год")
plt.ylabel("Количество фильмов")
plt.show()
```

Посчитаем **по каждому году** долю фильмов с указанной информацией о прокате в кинотеатрах, и без нее.

```python
films_part = []
for i, j in films_per_year.values():
    films_part.append(round(i / (i + j), 2))
films_part
```

```python
fig, ax = plt.subplots(figsize=(11, 4))         
ax.bar(years, films_part)
plt.xlabel("Год")
plt.ylabel("Доля фильмов с указанной \n информацией о прокате")
plt.title("Доля фильмов с указанной информацией о прокате в кинотеатрах по годам");
```

**Вывод**

В период с 2010 по 2014 год доля фильмов с информацией о прокате в кинотеатрах постепенно увеличивается. С 2015 года количество информации резко увеличивается с 28 до 65% и после этого значения колеблятся в пределах +-10%. Подобное поведение данных на графике вероятно связано с цифровизацией и автоматизацией сбора информации. Период с 2015 по 2019 годы представлен наиболее полно всего объема данных.

Введем 2 новый столбца:

- `budget_payback` - фильмы, кассовые сборы которых превысили затраченный на фильм бюджет

```python
data_supported['support_payback'] = (data_supported['box_office'] / data_supported['support']) > 1
```

```python
data_supported['budget_payback'] = (data_supported['box_office'] / data_supported['budget']) > 1
```

Теперь посмотрим, в каком соотношении распределяются фильмы, которые "окупились" и "не окупились", т.е. кассовые сборы которых превышают/не превышают общий размер выделенных на съемки средств.

```python
# Данные по оси X
years = sorted(list(data_supported['year'].unique()))

# Данные по оси Y
supported_films_per_year = {}

# Первое значение - кол-во фильмов, НЕ окупивших выделенный бюджет
# Второе значение - кол-во фильмов, окупивших выделенный бюджет
for year in sorted(data_supported['year'].unique()):
    supported_films_per_year[year] = [data_supported.loc[(data_supported['year'] == year) & (data_supported['budget_payback'] == False), 'year'].count(), \
                            data_supported.loc[(data_supported['year'] == year) & (data_supported['budget_payback'] == True), 'year'].count()]
                        
supported_films_counts = {
    'Кол-во фильмов, НЕ окупивших выделенный бюджет': np.array([i[0] for i in list(supported_films_per_year.values())]),
    'Кол-во фильмов, окупивших выделенный бюджет': np.array([i[1] for i in list(supported_films_per_year.values())]),
}
```

```python
# Строим график
fig, ax = plt.subplots(figsize=(12, 5))
bottom = np.zeros(len(years))
width = 0.6

for year, film_count in supported_films_counts.items():
    p = ax.bar(years, film_count, width, label=year, bottom=bottom)
    bottom += film_count

    ax.bar_label(p, label_type='center')

ax.set_title('Количество фильмов с государственной поддержкой по годам')
ax.legend()
plt.xlabel("Год")
plt.ylabel("Количество фильмов")
plt.show()
```

- Самый популярный жанр фильмов - драма. На 80 фильмов в этом жанре было выделено в общей сложности около 5 млрд. руб.
- В среднем, фильмы поддерживаемые финансово государством показывают невысокий средний рейтинг (6.0). Средний рейнинг фильмов по всей предоставленной нам выборке равен 6.6. При этом ни в одином из рассматриваемых нами периодов средний рейтинг фильмов за год не превышал этого значения. 
- Фильм с самым большим значением выделенных государством бюджетных средств на его поддержку - "Союз спасения" (А.Кравчук, 2019). На его поддержку было выделено 500 млн. руб. При этом рейтинг этого фильма составил те самые "средние" 6.00.
- Самый высокобюджетный фильм - "Тайна Печати Дракона" (О.Степченко, 2019). Его общий бюджет составил 2.3 млрд.руб., при государственной поддержке в 270 млн.руб. и кассовых сборах 333 млн.руб.
- Фильм с самым большим размером кассовых сборов - "Холоп" (К.Шипенко, 2019). За время проката он собрал более 3 млрд. руб. При этом рейтинг фильма нельзя назвать высоким - 6.80.
- Самым высокорейтинговым фильмом в данной подборке стал "Главный" (Ю.Кара, 2015). Пользовательский рейтинг этого фильма составил рекордные 9.4). Возможно фильм еще не набрал достаточное количество оценок и такое высокое значение не стоит воспринимать в серьез.
- Наконец больше всего бюджетных средств было выделено на поддержку фильмов Н. Хомерики - 660 млн. руб.
- Кореляция между популярностью фильма и его бюждетом/размером кассовых сбором - карйне низная и составила 0.10 и 0.15 соответственно.
